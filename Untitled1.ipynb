```python
from ipywidgets import widgets
from PIL import Image, ImageOps
from transformers import BlipProcessor, BlipForConditionalGeneration, MarianMTModel, MarianTokenizer
import torch
import nltk
import re
import io
from IPython.display import display

# Download NLTK Data
nltk.download('punkt')

# Load Models
print("Loading models... Please wait.")

# BLIP Model
blip_processor = BlipProcessor.from_pretrained("Salesforce/blip-image-captioning-base")
blip_model = BlipForConditionalGeneration.from_pretrained("Salesforce/blip-image-captioning-base")

# Translation Model
translation_model_name = 'Helsinki-NLP/opus-mt-en-ar'
translator_tokenizer = MarianTokenizer.from_pretrained(translation_model_name)
translator_model = MarianMTModel.from_pretrained(translation_model_name)

# Preprocessing functions
def clean_text(text):
    cleaned = re.sub(r"[^\w\s]", "", text)
    cleaned = re.sub(r"_", " ", cleaned)
    return cleaned.strip()

def translate_to_arabic(text):
    inputs = translator_tokenizer(text, return_tensors="pt", padding=True)
    translated = translator_model.generate(**inputs)
    translated_text = translator_tokenizer.batch_decode(translated, skip_special_tokens=True)[0]
    return translated_text

# Process each uploaded image
def process_images(change):
    uploaded_files = upload_widget.value
    if uploaded_files:
        for filename, file_info in uploaded_files.items():
            print(f"\nProcessing image: {filename}")

            content = file_info['content']
            image = Image.open(io.BytesIO(content))
            image = image.convert('RGB')
            image = ImageOps.fit(image, (220, 230), method=0, bleed=0.0, centering=(0.5, 0.5))

            display(image)

            # Generate Caption
            inputs = blip_processor(image, return_tensors="pt")
            output = blip_model.generate(**inputs)
            caption = blip_processor.decode(output[0], skip_special_tokens=True)
            caption = clean_text(caption)

            # Translate to Arabic
            translation = translate_to_arabic(caption)

            # Display Results
            print(f"Caption (English): {caption}")
            print(f"Caption (Arabic): {translation}")

# Upload Widget
upload_widget = widgets.FileUpload(accept='image/*', multiple=True)
upload_widget.observe(process_images, names='value')

# Display the widget
display(upload_widget)
```

```python
# Importing necessary libraries
from ipywidgets import widgets
from PIL import Image, ImageOps
from transformers import BlipProcessor, BlipForConditionalGeneration, Blip2Processor, Blip2ForConditionalGeneration, GitProcessor, GitForCausalLM, MarianMTModel, MarianTokenizer
import torch
import io
import re
import nltk
from IPython.display import display

# Download NLTK Data
nltk.download('punkt')

# --- Load Models ---
print("Loading models... Please wait...")

# Load BLIP model
blip_processor = BlipProcessor.from_pretrained("Salesforce/blip-image-captioning-base")
blip_model = BlipForConditionalGeneration.from_pretrained("Salesforce/blip-image-captioning-base")

# Load BLIP2 model
blip2_processor = Blip2Processor.from_pretrained("Salesforce/blip2-flan-t5-xl")
blip2_model = Blip2ForConditionalGeneration.from_pretrained("Salesforce/blip2-flan-t5-xl")

# Load GIT model
git_processor = GitProcessor.from_pretrained("microsoft/git-base")
git_model = GitForCausalLM.from_pretrained("microsoft/git-base")

# Load Flamingo model (if available)
# flamingo_processor = FlamingoProcessor.from_pretrained("Salesforce/flamingo-image-captioning-base")
# flamingo_model = FlamingoForConditionalGeneration.from_pretrained("Salesforce/flamingo-image-captioning-base")

# Load Translation models (English → Arabic and English → French)
translation_model_name_ar = 'Helsinki-NLP/opus-mt-en-ar'
translator_tokenizer_ar = MarianTokenizer.from_pretrained(translation_model_name_ar)
translator_model_ar = MarianMTModel.from_pretrained(translation_model_name_ar)

translation_model_name_fr = 'Helsinki-NLP/opus-mt-en-fr'
translator_tokenizer_fr = MarianTokenizer.from_pretrained(translation_model_name_fr)
translator_model_fr = MarianMTModel.from_pretrained(translation_model_name_fr)

print("Models loaded ✅")

# --- Functions ---
def clean_text(text):
    cleaned = re.sub(r"[^\w\s]", "", text)
    cleaned = re.sub(r"_", " ", cleaned)
    return cleaned.strip()

def translate_to_arabic(text):
    inputs = translator_tokenizer_ar(text, return_tensors="pt", padding=True)
    translated = translator_model_ar.generate(**inputs)
    translated_text = translator_tokenizer_ar.batch_decode(translated, skip_special_tokens=True)[0]
    return translated_text

def translate_to_french(text):
    inputs = translator_tokenizer_fr(text, return_tensors="pt", padding=True)
    translated = translator_model_fr.generate(**inputs)
    translated_text = translator_tokenizer_fr.batch_decode(translated, skip_special_tokens=True)[0]
    return translated_text

def generate_caption_blip(image):
    inputs = blip_processor(image, return_tensors="pt")
    out = blip_model.generate(**inputs)
    caption = blip_processor.decode(out[0], skip_special_tokens=True)
    return caption

def generate_caption_blip2(image):
    inputs = blip2_processor(image, return_tensors="pt")
    out = blip2_model.generate(**inputs, max_new_tokens=50)
    caption = blip2_processor.decode(out[0], skip_special_tokens=True)
    return caption

def generate_caption_git(image):
    inputs = git_processor(images=image, return_tensors="pt")
    generated_ids = git_model.generate(pixel_values=inputs.pixel_values, max_length=50)
    caption = git_processor.batch_decode(generated_ids, skip_special_tokens=True)[0]
    return caption

# --- Image Upload and Processing Function ---
def process_image(change):
    uploaded_file = upload_widget.value
    if uploaded_file:
        filename = list(uploaded_file.keys())[0]
        content = uploaded_file[filename]['content']

        image = Image.open(io.BytesIO(content)).convert('RGB')
        image = ImageOps.fit(image, (300, 300), method=0, bleed=0.0, centering=(0.5, 0.5))

        # Display Image
        display(image)

        # Generate Captions using different models
        caption_blip = clean_text(generate_caption_blip(image))
        caption_blip2 = clean_text(generate_caption_blip2(image))
        caption_git = clean_text(generate_caption_git(image))
        # Uncomment and use Flamingo if accessible
        # caption_flamingo = clean_text(generate_caption_flamingo(image))

        # Translate captions to Arabic and French
        translation_blip_ar = translate_to_arabic(caption_blip)
        translation_blip2_ar = translate_to_arabic(caption_blip2)
        translation_git_ar = translate_to_arabic(caption_git)
        # translation_flamingo_ar = translate_to_arabic(caption_flamingo)

        translation_blip_fr = translate_to_french(caption_blip)
        translation_blip2_fr = translate_to_french(caption_blip2)
        translation_git_fr = translate_to_french(caption_git)
        # translation_flamingo_fr = translate_to_french(caption_flamingo)

        # Display Captions and Translations
        print(f"\n📸 Caption (BLIP, English): {caption_blip}")
        print(f"🌍 Translation (BLIP, Arabic): {translation_blip_ar}")
        print(f"🌍 Translation (BLIP, French): {translation_blip_fr}")

        print(f"\n📸 Caption (BLIP-2, English): {caption_blip2}")
        print(f"🌍 Translation (BLIP-2, Arabic): {translation_blip2_ar}")
        print(f"🌍 Translation (BLIP-2, French): {translation_blip2_fr}")

        print(f"\n📸 Caption (GIT, English): {caption_git}")
        print(f"🌍 Translation (GIT, Arabic): {translation_git_ar}")
        print(f"🌍 Translation (GIT, French): {translation_git_fr}")

        # Uncomment and display Flamingo if available
        # print(f"\n📸 Caption (Flamingo, English): {caption_flamingo}")
        # print(f"🌍 Translation (Flamingo, Arabic): {translation_flamingo_ar}")
        # print(f"🌍 Translation (Flamingo, French): {translation_flamingo_fr}")

# --- Upload Widget ---
upload_widget = widgets.FileUpload(accept='image/*', multiple=False)
upload_widget.observe(process_image, names='value')

display(upload_widget)
```
